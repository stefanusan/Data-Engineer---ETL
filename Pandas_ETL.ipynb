###Transform

```python
assetCode =['BTC', 'ETH', 'XRP', 'LTC']


#converting open,high,low,close into GBP
#if currency belong to the assetCode list

df['open'] = df[['open','asset']].apply(lambda x : (float(x[0])*0.80) if x[1] in assetCode else np.nan, axis=1 )
df['high'] = df[['high','asset']].apply(lambda x : (float(x[0])*0.80) if x[1] in assetCode else np.nan, axis=1 )
df['low'] = df[['low','asset']].apply(lambda x : (float(x[0])*0.80) if x[1] in assetCode else np.nan, axis=1 )
df['close'] = df[['close','asset']].apply(lambda x : (float(x[0])*0.80) if x[1] in assetCode else np.nan, axis=1 )
```

```python
#dropping rows with null value
df.dropna(inplace=True)
df
```

```python
#reset the data index
df.reset_index(inplace=True, drop = True)
df.head()
```

```python
df.columns
```

```python
#drop columns
df.drop(labels=['slug','ranknow', 'volume', 'market', 'close_ratio', 'spread'], inplace =True, axis=1)
df.head()
```

###Load

```python
#connect function opens a connection to the SQLite data
conn = sqlite3.connect('session.db')
print(conn)
```

```python
#Drop  a table
try :
  conn.execute('DROP TABLE IF EXISTS "Crypto"')
except Exception as e :
  raise(e)
finally:
  print("Table Dropped")
```

```python
#Create a new table named as Crypto
try :
  conn.execute('''
      CREATE TABLE Crypto
      (ID           INTEGER PRIMARY KEY,
      ASSET         TEXT    NOT NULL,
      NAME          TEXT    NOT NULL,
      Date          datetime,
      Open          Float DEFAULT 0,
      High          Float DEFAULT 0,
      Low           Float DEFAULT 0,
      Close         Float DEFAULT 0);''')
  print('Table Created Successfully :D');

except Exception as e :
  print(str(e))
  print('Table creation failed :(')

finally:
  conn.close()
```

```python
#convert dataframe to list
df_list = df.values.tolist()
```

```python
#make a new connection
conn = sqlite3.connect('session.db')
print(conn)
```

```python
#make a cursor
cur = conn.cursor()
print(cur)
```

```python
#insert data to database
try :
  cur.executemany("INSERT INTO Crypto(ASSET, NAME, Date, Open, High, Low, Close) VALUES (?,?,?,?,?,?,?)", df_list)
  conn.commit()
  print('Data Inserted Successfully :D')

except Exception as e :
  print(str(e))
  print('Data insertion Failed :(')

finally :
  conn.close()
```
